```r
json_data <- fromJSON(file="tmp/all.json")
json_data
```

```r
train_test_split <- function(all.data) {
    all.data.length <- length(all.data)
    train.length <- as.integer(all.data.length * 0.8)
    train.indexes <- 1:train.length
    
    train <- all.data[train.indexes]
    target <- all.data[-train.indexes]
    
    output <- list(train=train, target=target)
    
    return(output)
}
```

```r
split <- train_test_split(json_data$data)
split$train
```

```r
options(repr.plot.width=15, repr.plot.height=6)
train.ts <- ts(split$train, start=c(2020,01,01),frequency=24*365)
```
