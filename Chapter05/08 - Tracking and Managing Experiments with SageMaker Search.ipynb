```python
import boto3
import pandas
import numpy as np
```

```python
tagKey = "Job name"
tagValue = "job"

search_params={
   "MaxResults": 10,
   "Resource": "TrainingJob",
   "SearchExpression": {
      "Filters": [
         {
            "Name": "TrainingJobName",
            "Operator": "Contains",
            "Value": tagValue
         }
      ]
   }
}
```

```python
sagemaker = boto3.client(service_name='sagemaker')
results = sagemaker.search(**search_params)

results
```

```python
for result in results['Results']:
    training_time = result['TrainingJob']['TrainingTimeInSeconds']
    job_name = result['TrainingJob']['TrainingJobName']
    print(f"{job_name} - {training_time}")
```
